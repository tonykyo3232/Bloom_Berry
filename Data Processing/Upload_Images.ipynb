```python
def cvt_2_base64(file_name):
    with open(file_name , "rb") as image_file :
        data = base64.b64encode(image_file.read())
    return data.decode('utf-8')
```

```python
def upload_image(file_name, file_type):
    
    result = cvt_2_base64("static/" + file_name + "." + file_type)

    today = date.today()

    new_image = {
        'name': file_name,
        'type': file_type,
        'content': result,
        'upload_time': str(today)
    }
    
    records.insert_one(new_image)
```
